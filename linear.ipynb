## Решение Задачи Классификации используя библиотеку sklearn и методы SVM и Log_Regression.

### Для этого импортирем нужные пакеты 

```python
from sklearn.svm import SVC
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
import numpy as np
```

### Сгенерируем и подготовим данные

```python
plt.figure(figsize=(8,8))
X, y = make_blobs(50, 2, centers=2, random_state=5, cluster_std=4)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')

xlim = (min(X[:,0])-2, max(X[:,0])+2)
ylim = (min(X[:,1])-2, max(X[:,1])+2)
xg = np.linspace(xlim[0], xlim[1], 60)
yg = np.linspace(ylim[0], ylim[1], 40)
xx, yy = np.meshgrid(xg, yg)
Xgrid = np.vstack([xx.ravel(), yy.ravel()]).T
```

### Обучим алгоритм и отрисуем результат

```python
plt.figure(figsize=(8,8))
model=SVC(kernel='rbf')
model.fit(X,y)
Z =  model.predict(Xgrid)
Z = Z.reshape(xx.shape)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')    
plt.contourf(xx,yy,Z, cmap='RdBu', alpha=0.5)
plt.xticks(fontsize=22,c='green')
plt.yticks(fontsize=22,c='green')
```

### Продлаем тоже самое с Логистической регрессией

```python
plt.figure(figsize=(8,8))
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X,y)
Z =  model.predict_proba(Xgrid)
Z = Z[:,1].reshape(xx.shape)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')  
plt.contour(xx,yy,Z,[0.5],linewidths=2, colors='black')
plt.xticks(fontsize=22,c='green')
plt.yticks(fontsize=22,c='green')







```

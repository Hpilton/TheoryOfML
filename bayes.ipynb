## Решение Задачи Классификации используя библиотеку sklearn и методы SVM и Log_Regression.

### Для этого импортирем нужные пакеты 

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
X, y = make_blobs(250, 2, centers=2, random_state=5, cluster_std=1)
```

### Сгенерируем и подготовим данные

```python
xlim = (min(X[:,0]), max(X[:,0]))
ylim = (min(X[:,1]), max(X[:,1]))

xg = np.linspace(xlim[0], xlim[1], 60)
yg = np.linspace(ylim[0], ylim[1], 40)
xx, yy = np.meshgrid(xg, yg)
Xgrid = np.vstack([xx.ravel(), yy.ravel()]).T
```

### Обучим алгоритм и отрисуем результат испольузуя различные Байесовские классификаторы

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA(n_components=2)
lda.fit(X,y)
Z =  lda.predict_proba(Xgrid)
Z = Z[:, 1].reshape(xx.shape)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')    
plt.contour(xx,yy,Z, [0.5], linewidths=2., colors='black')
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
Z = gnb.fit(X, y).predict_proba(Xgrid)
Z = Z[:, 1].reshape(xx.shape)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')    
plt.contour(xx,yy,Z, [0.5], linewidths=2., colors='black')
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

clf = QuadraticDiscriminantAnalysis()
clf.fit(X, y)
Z = clf.predict_proba(Xgrid)
Z = Z[:, 1].reshape(xx.shape)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')    
plt.contour(xx, yy, Z, [0.5], linewidths=2, colors='black')

```
